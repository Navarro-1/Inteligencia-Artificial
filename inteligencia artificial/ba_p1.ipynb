```python
def ba(grafo):
    V=list(grafo.keys())
    v1='a'
    S=v1
    Vp=[v1]
    Ep=[]
    yd=V.copy() 
    yd.remove(v1)
    flag=False
    s=[]
    while True:
        flag=False
        for x in S:
                
            for y in yd:
                if (grafo [x][y]>0) and (y not in Vp):
                    Ep.append((x,y))
                    Vp.append(y)
                    s.append(y)
                    yd.remove(y)
                    flag=True
            try:
                _=[yd.remove(y) for y in s]
            except:
                        pass
        if not flag:
            T=(Ep,Vp)
            print(T)
            return T
        S=s.copy()
        s=[]
T=ba(grafo)
T
```
